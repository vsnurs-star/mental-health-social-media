# **Key Insights from Mental Health and Social Media Dataset**

## 1. Screen Time and Stress
Participants with higher daily screen time (above 300 minutes) consistently report elevated stress levels. This suggests prolonged digital exposure may contribute to mental strain.

## 2. Sleep and Mental Health
Sleep hours show a strong inverse relationship with stress and anxiety. Individuals sleeping fewer than 6 hours tend to have higher anxiety scores, while those with 7â€“8 hours report better mood levels.

## 3. Gender Differences
Box plot analysis reveals that females exhibit higher median anxiety levels compared to males. This highlights potential gender-specific vulnerabilities in mental health outcomes.

## 4. Platform Engagement
Engagement ratios vary across platforms. Instagram users show higher positive interaction ratios, while Twitter users report more negative interactions. This suggests platform culture influences mental well-being.

## 5. Physical Activity and Mood
Physical activity minutes are positively correlated with mood levels. Participants engaging in at least 30 minutes of daily activity report significantly lower stress and higher mood scores.

## 6. Interaction Quality
Positive interactions (likes, supportive comments) improve mood levels even when screen time is high. Conversely, negative interactions amplify stress and anxiety regardless of usage duration.
